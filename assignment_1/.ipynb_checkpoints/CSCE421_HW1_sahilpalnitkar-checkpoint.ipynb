```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("Smarket.csv")
```

```python
print(dataframe.head())
```

```python
print(dataframe.shape)
```

```python
X = dataframe[['Lag1','Lag2']]
y = dataframe['Direction']
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
plot_y = []
plot_x = []
for i in range(1,11):
    plot_y.append(i)
    print(str(i) + ":")
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train,y_train) #Trains model
    score_test = neigh.score(X_test, y_test) #.score uses .predict as an underlying function
    plot_x.append(score_test)
    print("test_" + str(i) + ":")
    print(score_test)
    
```

```python
plt.plot(plot_x)
plt.ylabel(plot_y)
plt.show()
```
